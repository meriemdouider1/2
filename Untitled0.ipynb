```
def creation_data(Data, RESIZE, sigmaX=10):
    IMG = []
    read = lambda imname: np.asarray(Image.open(imname).convert("RGB"))
    for IMAGE_NAME in tqdm(os.listdir(Data)):
        PATH = os.path.join(Data,IMAGE_NAME)
        _, ftype = os.path.splitext(PATH)
        if ftype == ".jpg":
            img = read(PATH)
            img = cv2.resize(img, (RESIZE,RESIZE))
           
            IMG.append(np.array(img))
    return IMG
```

# **Transfer learning**

Qu'est-ce que l'apprentissage par transfert?

L'apprentissage par transfert est une approche utilise dans l'apprentissage automatique dans lequel un modle cr et form pour une tche est rutilis comme point de dpart d'une tche secondaire. L'apprentissage par transfert diffre de l'apprentissage automatique traditionnel car il implique l'utilisation d'un modle pr-entran comme tremplin pour dmarrer une tche secondaire.

Cette approche imite la faon dont les humains appliquent les connaissances acquises pour une tche  une nouvelle tche. Par exemple, John Doe a appris  lire en premire anne. En dixime anne, John a utilis ses capacits de lecture en classe de chimie. Les connaissances qu'il avait acquises de la tche principale (apprendre  lire) sont devenues la base sur laquelle il a commenc la tche secondaire (apprendre la chimie).



Les avantages de l'utilisation de l'apprentissage par transfert et de l'apprentissage profond :

Prendre un modle qui a dj t form dans un domaine spcifique et le rappliquer  un autre domaine prsente de nombreux avantages. Certains des principaux avantages sont numrs ci-dessous.

- Moins de donnes d'entranement: commencer  entraner un modle  partir de zro demande beaucoup de travail et ncessite beaucoup de donnes. Par exemple, si nous voulons crer un nouvel algorithme capable de dtecter un froncement de sourcils, nous avons besoin de beaucoup de donnes d'entranement. Notre modle devra d'abord apprendre  dtecter les visages, et ce n'est qu'alors qu'il pourra apprendre  dtecter des expressions, telles que les froncements de sourcils. Au lieu de cela, si nous utilisons un modle qui a dj appris  dtecter les visages et que nous recyclons ce modle pour dtecter les froncements de sourcils, nous pouvons obtenir le mme rsultat en utilisant beaucoup moins de donnes.
Les modles se gnralisent mieux 
- l'utilisation de l'apprentissage par transfert sur un modle prpare le modle  bien fonctionner avec des donnes sur lesquelles il n'a pas t form. C'est ce qu'on appelle la gnralisation. Les modles qui ont t forms  l'aide de l'apprentissage par transfert sont mieux  mme de gnraliser d'une tche  une autre car ils ont t forms pour apprendre  identifier les fonctionnalits qui peuvent tre appliques  de nouveaux contextes.
Rend l'apprentissage en profondeur plus accessible 
- l'utilisation de l'apprentissage par transfert facilite l'utilisation de l'apprentissage en profondeur. Il est possible dobtenir les rsultats souhaits sans tre un expert en deep learning, en utilisant un modle cr par un spcialiste du deep learning et en lappliquant  un nouveau problme.

```
DATAGlua_Train = '/content/Fundus_Train_Val_Data/Fundus_Scanes_Sorted/Train'
DATAGlua_Test = '/content/Fundus_Train_Val_Data/Fundus_Scanes_Sorted/Validation'
```

notre jeu deux de donnes est divis en deux partie :

*   des donnes avec un Gluacoma_ngative



```
CATEGORIES  = list()
for root, dirs, files in os.walk(DATAGlua_Train, topdown=False):
    for name in dirs:
        CATEGORIES.append(os.path.join(name))
print (CATEGORIES)
```


Division l'ensemble de donnes d'entranement en x_train_data et y_train_data  : (x_train_data : 'les images' , y_train_data : 'label / catgorie ')

```
x_train_data = []
y_train_data = []
IMG_SIZE=224
def creation_train_data():
    for category in CATEGORIES:
      path = os.path.join(DATAGlua_Train, category)
      class_num = CATEGORIES.index(category)
      for img in os.listdir(path):
        try:
            img_array = cv2.imread(os.path.join(path, img), cv2.IMREAD_COLOR)
            new_array = cv2.resize(img_array, (IMG_SIZE,IMG_SIZE))
            x_train_data.append(new_array)
            y_train_data.append(class_num)
        except Exception as e:
            pass
creation_train_data()
```

Division l'ensemble de donnes test en x_test_data et y_test_data  : (x_test_data : 'les images' , y_test_data : 'label / catgorie ')




```
x_test_data = []
y_test_data = []
IMG_SIZE=224
def creation_test_data():
    for category in CATEGORIES:
      path = os.path.join(DATAGlua_Test, category)
      class_num = CATEGORIES.index(category)
      for img in os.listdir(path):
        try:
            img_array = cv2.imread(os.path.join(path, img), cv2.IMREAD_COLOR)
            new_array = cv2.resize(img_array, (IMG_SIZE,IMG_SIZE))
            x_test_data.append(new_array)
            y_test_data.append(class_num)
        except Exception as e:
            pass
creation_test_data ()
```

```
X_Train, X_Val, Y_Train, Y_Val = train_test_split(x_train_data, y_train_data, test_size=0.05, random_state=1)
```

# **Cration du modle**: 


Qu'est-ce qu'un modle pr-entran ?
Un modle pr-entran a t pralablement form sur un ensemble de donnes et contient les pondrations et les biais qui reprsentent les caractristiques de l'ensemble de donnes sur lequel il a t form. Les fonctionnalits apprises sont souvent transfrables  diffrentes donnes. Par exemple, un modle form sur un grand ensemble de donnes d'images d'oiseaux contiendra des caractristiques apprises telles que des artes ou des lignes horizontales que vous seriez transfrables  votre ensemble de donnes.

```
from tensorflow.keras.applications.inception_v3 import InceptionV3
```

```
IMG_SHAPE = (224, 224,3)
```

```
#datagen= ImageDataGenerateur(rescale=1.0/255)
base_model1 = InceptionV3(include_top=False,
                   input_shape=IMG_SHAPE,
                   pooling='avg',classes=2,
                   weights='imagenet')
```

```
for layer in base_model1.layers:
    print(layer.name)
    layer.trainable = False
    
print(len(base_model1.layers))
```

```
base_model1.summary()
```

```
base_model1.trainable = False
```

```
from tqdm import tqdm_notebook
from keras.preprocessing.image import ImageDataGenerator
from keras.layers import Input, Lambda, Dense, Flatten
from keras.models import load_model, Model
```

```
# Flatten the output layer to 1 dimension
x = base_model1.output
# Ajouter une couche entirement connecte avec 512 units caches et l'activation ReLU
x = layers.Dense(512, activation='relu')(x)
# Ajouter un taux d'abandon de 0,5 (droupout)
x = layers.Dropout(0.5)(x)
# Ajouter une dernire couche sigmode pour la classification
x = layers.Dense(2, activation='softmax')(x)
```

```
# Configurer et compiler le modle
model1 = Model(base_model1.input, x)
optimizer = Adam(lr=0.0001, beta_1=0.9, beta_2=0.999, epsilon=None, decay=0.0, amsgrad=True)
model1.compile(loss='categorical_crossentropy',
              optimizer=optimizer,
              metrics=['accuracy'])
```
